*   Data from Macromolecules 2014, 47, 6453−6461 (dx.doi.org/10.1021/ma500923r)


```python
import utils
import scipy.constants as sc
import numpy as np

temperature = 700  # K
kTval = sc.R * temperature / 1000  # kJ/mol
# molecular chain
# T-bond-DPP-bond-T-bond-T-bond-E-bond-T-bond
l = [2.527, 1.438]
# change the logic of bond order and angle order，l_array[1] rotated by angle[1]
Angle = np.deg2rad(np.array([-15, -15]))
rotation = np.array([0, 1])
labels = {
    1: {'label': 'thiophene-thiophene', 'color': 'b'},
    # 2: {'label': 'dihedralB', 'color': 'm'},
    # 3: {'label': 'T-E', 'color': 'c'},
    # 4: {"label": "FT-FT", "color": 'g'}, 
    # 5: {"label": "ADTDI-FT", "color": 'r'},
}
```

```python
all_data = {}
for key, label_info in labels.items():
    all_data[key] = {**utils.setup_interpolation(label_info, kTval), **label_info}
utils.plot_dihedral_potentials(all_data)
```

```python
lp_repeats, lam = utils.compute_correlation(all_data=all_data,
                                                       l_list=l,
                                                       Angle_rad=Angle,
                                                       rotation_types=rotation,
                                                       kTval=kTval)

print(f"Max eigen value: lambda_max = {lam:.12f}")
print(f"Correlation length = {lp_repeats:.6f}")
# Monte Carlo sampling:
# 10.3279 (dx.doi.org/10.1021/ma500923r supporting information)
# 10.382150311850822 (https://github.com/Swordshinehjy/DFT_persistence_length_Monte_Carlo_sampling/blob/main/P3HT.ipynb)
```
